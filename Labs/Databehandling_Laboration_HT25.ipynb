## Uppgift (a)

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

# Läs in datafilen College.csv
college = pd.read_csv("College.csv")
college.head()
```

## Uppgift (b)

```python
# Läs in filen direkt med index_col=0
college2 = pd.read_csv('College.csv', index_col=0)
college2.head()
```

```python
# Gör det manuellt
college3 = college.rename({'Unnamed: 0': 'College'}, axis=1)
college3 = college3.set_index('College')
college3.head()
```

## Uppgift (c)

```python
college.describe()
```

## Uppgift (d)

```python
cols = ['Top10perc', 'Apps', 'Enroll']
scatter_matrix(college[cols], figsize=(8, 8), alpha=0.8)
```

## Uppgift (e)

```python
college.boxplot(column='Outstate', by='Private', figsize=(6,6))
plt.title('Out-of-state Tuition by Private/Public')
plt.suptitle('')
plt.xlabel('Private')
plt.ylabel('Outstate Tuition')
plt.show()
```

## Uppgift (f)

```python
college['Elite'] = pd.cut(college['Top10perc'], [0, 50, 100], labels=['No', 'Yes'])
college['Elite'].value_counts()
```

```python
college.boxplot(column='Outstate', by='Elite', figsize=(6,6))
plt.title('Out-of-state tuition för Elite och icke-Elite universitet')
plt.suptitle('')
plt.xlabel('Elite-status')
plt.ylabel('Outstate tuition')
plt.show()
```

## Uppgift (g)

```python
# Välj några kvantitative variabler att undersöka
vars_to_plot = ['Apps', 'Accept', 'Enroll', 'Expend']

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Loopa genom variablerna och skapa histogram
for i, var in enumerate(vars_to_plot):
    ax = axes[i // 2, i % 2] # Placera varje plot i rätt ruta
    college[var].plot.hist(bins=20, ax=ax, edgecolor='black')
    ax.set_title(f'Fördeling av {var}')
    ax.set_xlabel(var)
    ax.set_ylabel('Antal universitet')
    
plt.tight_layout()
plt.show()
```

```python
plt.Figure(figsize=(6,4))
college['Expend'].plot.hist(bins=10, alpha=0.6, label='10 bins')
college['Expend'].plot.hist(bins=30, alpha=0.6, label='30 bins')
plt.title('Jämförelse av olika antal bins för Expend')
plt.xlabel('Utbildningskostnad per student ($)')
plt.ylabel('Antal universitet')
plt.legend()
plt.show()
```

## Uppgift (h)

```python
# Undersöka genomsnittliga värden beroende på Elite-status
college[['Personal', 'Expend', 'Outstate', 'Elite']].groupby('Elite').mean()
```

```python
# Undersöka korrelationer mellan några numeriska variabler
corr = college[['Personal', 'Expend', 'Top10perc', 'Outstate']].corr()
corr
```

```python
# Visualisera sambandet mellan Expend och Top10perc
college.plot.scatter(x='Expend', y='Top10perc', alpha=0.6)
plt.title('Samband mellan undervisningskostnad och andel toppstudenter')
plt.xlabel('Utbildningskostnad per student ($)')
plt.ylabel('Andel studenter från top 10%')
plt.show()
```

```python
college.boxplot(column='Personal', by='Elite', figsize=(6,6))
plt.title('Personalkostnader beroende på Elite-status')
plt.suptitle('')
plt.xlabel('Elite-status')
plt.ylabel('Personalkostnader ($)')
plt.show()
```

I denna del utforskar jag datan ytterligare för att se samband mellan olika kostnadsvariabler och universitetens status som "Elite".

Medelvärden visar att elituniversitet (Elite = Yes) tenderar att spendera mer per student, både i undervisning (Expend) och personal (Personal).
Boxplotten bekräftar att dessa skolor generellt har högre kostnader och ofta även högre "Outstate tuition".

Korrelationstabellen visar tydlig positiv samband mellan:
* Expend och Top10perc (ju mer man spenderar per student, desto större andel toppstudenter)
* Expend och Outstate (dyrare skolor tenderar att vara mer selektiva och exklusiva)

Scatterplotten mellan Expend och Top10perc visar ett stigande mönster, vilket tyder på att högre investering i utbildning ofta ger en högre andel högpresterande studenter.

Slutsats:
Universitet med högre ekonomiska resurser har en tendens att locka mer konkurrenskraftiga studenter.
Detta visar en tidlig koppling mellan utbildningskvalitet och ekonomiska investeringar.
